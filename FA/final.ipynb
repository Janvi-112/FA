```python
df = s.copy()
df1 = df.iloc[:, :10]
df2 = df.iloc[:, 11:]
df2 = df2.fillna(0)
```

```python
normalized_df2=(df2-df2.min())/(df2.max()-df2.min())
```

```python
normalized_df2.describe()
```

```python
normalized_df = pd.concat([df1, normalized_df2], axis=1)
```

```python
a = 3
b = 2
c = 1
d = -1
e = -2
f = -3
```

```python
df = normalized_df.copy()
```

```python
df['goalkeeper'] = (f*normalized_df.GA90 + b*normalized_df.Save1 + b*normalized_df.W + c*normalized_df.D 
              + d*normalized_df.L + a*normalized_df.Save1 + e*normalized_df.PSxG1 + b*normalized_df.Cmp 
              + c*normalized_df.Stp1 + b*normalized_df.OPA90) / (2*a + 5*b + 3*c)

df['defender'] = (c*normalized_df.Gls + b*normalized_df.Ast + e*normalized_df.CrdY + f*normalized_df.CrdR 
            + c*normalized_df.Cmp1 + f*normalized_df.Err + f*normalized_df.CrdY2 + b*normalized_df.Fls
            + b*normalized_df.PrgDist + c*normalized_df.KP + c*normalized_df.Recov
            + c*normalized_df.CrsPA + c*normalized_df.SCA90 + c*normalized_df.GCA90 + b*normalized_df.TklW
            + b*normalized_df.Tkl1 + f*normalized_df.Past + c*normalized_df.Succ1
            + f*normalized_df.PKcon + c*normalized_df.Won1) / (5*a + 6*b + 9*c)

df['forward'] = (c*normalized_df.Gls + b*normalized_df.Ast + e*normalized_df.CrdY + f*normalized_df.CrdR 
           + b*normalized_df.SoT1 + b*normalized_df.SoT90 + c*normalized_df.GSh + c*normalized_df.GxG 
           + a*normalized_df.npGxG + c*normalized_df.Cmp1 + b*normalized_df.AxA + b*normalized_df.KP 
           + b*normalized_df.Third + a*normalized_df.PPA + c*normalized_df.Prog + a*normalized_df.SCA90 
           + a*normalized_df.GCA90 + c*normalized_df.AttPen + c*normalized_df.Succ11 + a*normalized_df.CPA 
           + e*normalized_df.Mis + f*normalized_df.CrdY2 + d*normalized_df.Off) / (7*a + 7*b + 9*c)


df['midfielder'] = (c*normalized_df.Gls + b*normalized_df.Ast + e*normalized_df.CrdY + f*normalized_df.CrdR 
              + b*normalized_df.SoT1 + c*normalized_df.GSoT + c*normalized_df.npGxG + c*normalized_df.Cmp1 
              + b*normalized_df.AxA + c*normalized_df.KP + a*normalized_df.Third + a*normalized_df.PPA 
              + b*normalized_df.Prog + d*normalized_df.Int + d*normalized_df.Blocks + a*normalized_df.SCA90 
              + a*normalized_df.GCA90 + b*normalized_df.Mid3rd + e*normalized_df.Past + c*normalized_df.Succ1 
              + c*normalized_df.Mid3rd11 + b*normalized_df.Succ11 + b*normalized_df.PrgDist + a*normalized_df.CPA
              + f*normalized_df.Mis + e*normalized_df.Dis + f*normalized_df.CrdY2 
              + d*normalized_df.Fls) / (8*a + 10*b + 10*c)
```
